```python
"""ADD NEW FEATURES TO DATASET
ride_length: DESCRIBES LENGTH OF TIME OF THE RIDE
week_day: DAY OF THE WEEK
day_of_week : THE DAY OF THE WEEK the bike ride started"""
import datetime as dt
from datetime import datetime, timedelta
total_df['started_at'] = pd.to_datetime(total_df['started_at']) #convert started at and ended at to datetime objects
total_df['ended_at'] = pd.to_datetime(total_df['ended_at'])
total_df['ride_length'] = total_df['ended_at'] - total_df['started_at']

total_df['started_at'] = pd.to_datetime(total_df['started_at'])
total_df['day_of_week'] = total_df['started_at'].apply(lambda x : x.isoweekday())
total_df.head(24)
```

```python
#NEXT PROCESSING STEPS 
"""FIGURE OUT WHAT TO DO WITH NEGATIVE TIME DELTAS OF TIME_LENGTH"""
#dropped those rows
"""FIGURE OUT WHAT TO DO WITH NA VALUES WITH START STATION NAME AND ID
   - SEE IF YOU CAN DETERMINE THE START STATION NAME AND ID BASED ON START LAT AND LNG """
#Imputed 312788 nan values with matching start Lat and Lng start station rows
"""FIGURE OUT WHAT TO DO WITH NA VALUES WITH END STATION NAME AND ID
   - SEE IF YOU CAN DETERMINE THE END STATION NAME AND ID BASED ON END LAT AND LNG"""
#Imputed 291544 nan values with matching end lat and lng end station rows
"""FIGURE OUT WHAT TO DO WITH NA VALUES WITH END LAT AND LNG
   - IF END STATION NAME AND END STATION ID ARE ALSO NA IN THE SAME ROW JUST DROP THAT ROW"""

"""CREATE A NEW COLUMN CALLED TRIP DISTANCE THAT DETERMINES HOW FAR THE TRIP WAS FROM THE START STATION TO END STATION"""
```

```python
"""FIGURE OUT WHAT TO DO WITH NEGATIVE TIME DELTAS OF TIME_LENGTH"""
#DISCARD ANY ROWS WITH NEGATIVE TIME DELTA VALUES FOR TIME_LENGTH COLUMNN
#THESE ROWS CONTAIN NON SIGNIFICANT INFORMATION AND ARE LIKELY IMPROPER DATA ENTRIES OR ENTRYS OF DATA OF RIDERS TAKING A BIKE
#AND IMMEDIATELY PUTTING THE BIKE BACK 
#MAJORITY OF THESE NEGATIVE TIME DELTAS FALL WITHIN AN HOUR UPON BIKE PURCHASE
time = dt.timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)
total_df = total_df[total_df['ride_length'] >= time] 
total_df.shape
```

```python
import seaborn as sns
# Convert 'ride_length' from timedelta to minutes for easier visualization
new_df['ride_length_in_minutes'] = new_df['ride_length'].dt.total_seconds() / 60

# Create separate dataframes for casual and member users
casual_df = new_df[new_df['member_casual'] == 'casual']
member_df = new_df[new_df['member_casual'] == 'member']

casual_df = casual_df['ride_length_in_minutes']
casual_df.clip(0, 100, inplace = True)
member_df = member_df['ride_length_in_minutes']
member_df.clip(0, 100, inplace = True)
# Plotting the distribution of ride lengths for both groups

plt.hist(casual_df, color='blue', label='Casual')
plt.hist(member_df, color='green', label='Member')

plt.xlabel('Ride Length (in minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Ride Lengths for Casual and Member Users')
plt.legend()
plt.show()
```
